```python
from nltk.tokenize import word_tokenize, sent_tokenize
```

```python
total_corpus = pd.read_csv('total_corpus.csv')
```

```python
import swifter
```

```python
def process_disease_drug_rows(x):
    disease = x['disease']
    drug = x['drug']
    articles = x['articles'].split(',')
    for article in articles:
        content = str(total_corpus[total_corpus['cord_uid'] == article].full_text.values[0])
        chosen = []
        list1 = [str(drug).lower(), str(disease).lower()]
        for sentence in sent_tokenize(content):
            l = word_tokenize(sentence.lower())
            if len(set(l).intersection(list1)) > 1:
                chosen.append(sentence)
        if len(chosen) > 0:
            sentences = ''.join(chosen)
            return disease, drug, sentences
        else:
            return disease, drug, None
```

```python
new_df = drug_disease_list.swifter.apply(process_disease_drug_rows, axis=1, result_type='expand')
```

```python
new_df.describe()
```

```python
new_df.columns = ['disease', 'drug', 'sentences']
```

```python
new_df[new_df['sentences'].notna()]
```

```python
drug_disease_sentences = new_df[new_df['sentences'].notna()]
```
