```python
metadata = pd.read_csv('../CORD19/metadata.csv')
```

```python
metadata = metadata[['title', 'doi', 'abstract']]
```

```python
metadata['metaId'] = range(1, len(metadata) + 1)
```

```python
pdb_ids = pd.read_csv('../August/Dictionary/Final_PDB_2011_11_03.txt', header=None)
```

```python
pdb_ids.columns = ['pdb']
```

```python
metadata = metadata[metadata.abstract.notna()]
```

```python
len(metadata)/2
```

```python
abstract_list_1 = ' '.join(metadata[0:284512]['abstract'].tolist())
abstract_list_2 = ' '.join(metadata[284513:]['abstract'].tolist())
```

```python
ts = TextSearch(case='ignore', returns='match')
ts.add(pdb_ids.pdb.tolist())
found_pdbs_1 = ts.findall(abstract_list_1)
```

```python
found_pdbs_2 = ts.findall(abstract_list_2)
```

```python
found_pdbs_2.extend(found_pdbs_1)
```

```python
metadata['Keyword1'] = metadata['abstract'].apply(lambda x: ', '.join(set(x.split()).intersection(set(unique_drugs_df.drug.tolist()))))
```

```python
metadata['Keyword2'] = metadata['abstract'].apply(lambda x: ', '.join(set(x.split()).intersection(set(unique_pdbs_df.pdb.tolist()))))
```

```python
found_pdb_drug = metadata[(metadata.Keyword1 != '') & (metadata.Keyword2 != '')]
```

```python
drug_pdb_abstract_map = found_pdb_drug.assign(Keyword1=found_pdb_drug.Keyword1.str.split(',')).explode(['Keyword1'])
```

```python
drug_pdb_abstract_map = drug_pdb_abstract_map.assign(Keyword2=found_pdb_drug.Keyword2.str.split(',')).explode(['Keyword2'])
```
