# Step 4: Map disease names to articles

```python
import pandas as pd
from tqdm import tqdm
```

```python
total_corpus = pd.read_csv('total_corpus.csv')
```

```python
total_corpus.head()
```

```python
disease_df = pd.read_csv('disease_found_list.csv')
```

```python
disease_df.tail()
```

```python
disease_list = disease_df[disease_df['disease'] != 'disease']['disease'].tolist()
disease_list[0:10]
```

```python
disease_articles_list = []
```

```python
for disease in tqdm(disease_list):
    _list = ','.join(total_corpus[total_corpus['full_text'].str.contains(disease, case=False, regex=False, na=False)]['cord_uid'].tolist())
    if len(_list) > 0:
        disease_articles_list.append([disease, _list])
```

```python
disease_articles_list[0:10]
```

```python
disease_articles_list_df = pd.DataFrame(disease_articles_list, columns=['disease', 'article'])
```

```python
disease_articles_list_df.to_csv('disease_article_list.csv', index=False)
```
